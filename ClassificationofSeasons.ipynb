```python
import pandas as pd
import numpy as np
```

```python
pitchingall = pd.read_csv('PitchersDataByYear.csv')
positionsall = pd.read_csv('PlayerDataByYear.csv')
```

```python
pitchingall
```

```python
positionsall
```

```python
positions = positionsall[positionsall.G.ge(100)]
```

```python
positions
```

```python
pitching = pitchingall[pitchingall.G.ge(15)]
```

```python
pitching
```

```python
positions = positions[["Season", 'WAR', 'playerid']] 
pitching = pitching[["Season", 'WAR', 'playerid']]
```

```python
pitching.head()
```

```python
frames = [pitching, positions]
```

```python
allwar = pd.concat(frames)
```

```python
allwar
```

```python
bins = [-4, 0, 1, 2, 3, 4, 5, 6,np.inf]
names = ['worst', 'bad', 'poor', 'average', 'above average', 'good', 'great', 'superstar']
```

```python
allwar['seasonclass'] = pd.cut(allwar['WAR'], bins, labels = names)
```

```python
allwar
```

```python
allwar['seasonclass'].count()
```

```python
allwar.groupby('seasonclass').seasonclass.count() # this is split by season, not a cumulative war for each player but 
#classifying each season played. 
```

This is filtereed by players with over a 100 games in one season and pitchers over 15 games in one season. These are not hard set filters, just for showing the example.
