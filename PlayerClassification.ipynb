```python
import pandas as pd
import numpy as np
```

```python
pitching = pd.read_csv('PitchersDataByYear.csv')
positions = pd.read_csv('PlayerDataByYear.csv')
```

```python
pitching.head()
```

```python
positions
```

```python
positions = positions[["Season", 'WAR', 'playerid']]
pitching = pitching[["Season", 'WAR', 'playerid']]
```

```python
pitching.head()
```

```python
frames = [pitching, positions]
```

```python
allwar = pd.concat(frames)
```

```python
allwar
```

```python
bins = [-4, 0, 1, 2, 3, 4, 5, 6,np.inf]
names = ['worst', 'bad', 'poor', 'average', 'above average', 'good', 'great', 'superstar']
```

```python
allwar['playerclass'] = pd.cut(allwar['WAR'], bins, labels = names)
```

```python
allwar
```

```python
allwar['playerclass'].count()
```

```python
allwar.groupby('playerclass').playerclass.count() # does not filter out players that barely played at all, IP ETC
```

The abundance of players in worst probably is due to a large amount of players who have barely played enough to accumulate much WAR at all. This will have to be filtered out initially.
